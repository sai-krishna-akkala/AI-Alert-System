```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")  # base model

model.train(
    data="/content/people_counting-1/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10,
    optimizer="Adam",
    lr0=0.001,
    device=0
)
```

```python
model.val()
```

```python
model.predict(source="/content/crowd_c.jpg", save=True)
```

```python
results = model("/content/crowd_c.jpg")

for r in results:
    print("People count:", len(r.boxes))
```
